```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd.options.display.max_columns = None
```

```python
data = pd.read_csv('accurate_data_1018.csv')
data_inacc = pd.read_csv('inacc_data_1018.csv')
```

```python
to_add = data_inacc[data_inacc.nodes_per_layer > 20][data_inacc.method == 'adversarial'][data_inacc.optimizer == 'adam']
```

```python
# merge data and to_add into one dataframe 
data = data.append(to_add)
print("len data: ", len(data))
# reset the index of data
data = data.reset_index(drop=True)
```

```python
# assign softplus, standard, and adversarial to colors
colors = {'softplus': 'red', 'standard': 'blue', 'adversarial': 'green'}
for optimizer in ['sgd', 'adam']:
    for fixed_seed in [1.0, 0.0]:
        for nodes_per_layer in [25, 50, 75, 100]:
            for num_layers in [3, 5]:
                subset = data[data.optimizer == optimizer][data.fixed_seed == fixed_seed][data.nodes_per_layer == nodes_per_layer][data.num_layers == num_layers]
                if len(subset) == 0:
                    continue
                plt.xlabel("data perturbation threshold")
                for metric in ['top1', 'top3', 'top5', 'gradnorm']:
                    plt.title(f"Opt: {optimizer}, Fixed Seed: {fixed_seed}, Nodes: {nodes_per_layer}, Layers: {num_layers}, Metric: {metric}")
                    plt.scatter(subset.threshold, subset[metric], label=metric, color = subset.method.map(colors))
                    plt.show()
```

```python
# assign softplus, standard, and adversarial to colors
colors = {'softplus': 'red', 'standard': 'blue', 'adversarial': 'green'}
for optimizer in ['sgd', 'adam']:
    for fixed_seed in [1.0, 0.0]:
        for nodes_per_layer in [25, 50, 75, 100]:
            for num_layers in [3, 5]:
                subset = data[data.optimizer == optimizer][data.fixed_seed == fixed_seed][data.nodes_per_layer == nodes_per_layer][data.num_layers == num_layers]
                if len(subset) == 0:
                    continue
                plt.xlabel("data perturbation threshold")
                for metric in ['top1', 'top3', 'top5', 'gradnorm']:
                    plt.title(f"Opt: {optimizer}, Fixed Seed: {fixed_seed}, Nodes: {nodes_per_layer}, Layers: {num_layers}, Metric: {metric}")
                    m25, m75 = metric + "_25", metric + "_75"
                    i = -0.01
                    for method in ['softplus', 'standard', 'adversarial']:
                        i += 0.005
                        subset_method = subset[subset.method == method]
                        if len(subset_method) == 0:
                            continue
                        error = [subset_method[metric] - subset_method[m25], subset_method[m75] - subset_method[metric]]                        
                        plt.errorbar(subset_method.threshold.astype(float) + i, subset_method[metric], yerr=error, fmt='o', color = colors[method])
                    plt.show()
```

From the graphs, it seems like softplus (red) is generally best -- but sometimes (especially for the gradnorm metric) adversarial is better. 
This trend is consistent but not always significant -- the middle 50% range (error bars) has overlap for many of the graphs.

Next, we'll try to look at over vs. underfitting.
We will use train_acc - test_acc as a proxy for underfitting (high score means overfit, lower score underfit)

```python
data.head()
```

```python
test_acc = np.array(data.avg_acc_test)
train_acc = np.array(data.avg_acc_train)
overfitting = train_acc - test_acc
data['overfitting'] = overfitting
```

```python
for metric in ['gradnorm', 'top1', 'top3', 'top5']:
    plt.scatter(data.overfitting, data[metric], color = data.method.map(colors))
    plt.xlabel("overfitting index")
    plt.ylabel(metric)
    plt.title("Overfitting/robustness by method")
    plt.show()
```

Seems clear that adversarial is underfit. But the degree of overfitting doesn't obviously correlate with robustness (it still might if we look at specific parameters, though). 
Next, I'll narrow in on just softplus and regular to see if there is any pattern that is more obvious with less data on the chart.

```python
data_not_adv = data[data.method == 'softplus'].append(data[data.method == 'standard'])
for metric in ['gradnorm', 'top1', 'top3', 'top5']:
    plt.scatter(data_not_adv.overfitting, data_not_adv[metric], color = data_not_adv.method.map(colors))
    plt.xlabel("overfitting index")
    plt.ylabel(metric)
    plt.title("Overfitting/robustness by method")
    plt.show()
```

The above graphs indicate that softplus is less likely -- compared with regular training -- to be overfit. This makes sense! 

```python
for optimizer in ['adam', 'sgd']:
    data_lim = data[data.optimizer == optimizer]
    for metric in ['gradnorm', 'top1', 'top3', 'top5']:
        plt.scatter(data_lim.overfitting, data_lim[metric], color = data_lim.method.map(colors))
        plt.xlabel("overfitting index")
        plt.ylabel(metric)
        plt.title("Overfitting/robustness for " + optimizer)
        plt.show()
optimizers = {'adam': 'red', 'sgd': 'blue'}
for method in ['adversarial', 'standard', 'softplus']:
    plt.title(method + ": overfitting vs gradnorm, by optimizer")
    plt.scatter(data[data.method == method].overfitting, data[data.method == method].gradnorm, color = data[data.method == method].optimizer.map(optimizers))
    plt.show()
```

Interesting. Seems to be that adam leads to much more underfitting for adversarial (but could this just be an artifact of what data is available since generally SGD's accuracy was too low to include?).
And for the other methods, adam slightly overfit but was very clustered and SGD has a wider range of values. Also, SGD's gradnorm values are lower, typically -- is this real or just an artifact of what data was accurate-enough?
